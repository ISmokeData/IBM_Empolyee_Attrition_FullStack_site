### **Data**
Hereâ€™s a brief explanation of the columns in the **Attrition** dataset:

1. **Age**: The age of the employee.
2. **Attrition**: Whether the employee has left the company (Yes/No).
3. **BusinessTravel**: Frequency of employee's business travel (e.g., Non-Travel, Travel_Rarely, Travel_Frequently).
4. **DailyRate**: Daily salary of the employee.
5. **Department**: The department the employee works in (e.g., Sales, R&D, HR).
6. **DistanceFromHome**: The distance between the employee's home and workplace.
7. **Education**: Education level (numerical, typically from 1 to 5).
8. **EducationField**: The field of education (e.g., Life Sciences, Marketing).
9. **EmployeeCount**: Typically a constant column with the value 1 (used for counting employees).
10. **EmployeeNumber**: Unique identifier for each employee.
11. **EnvironmentSatisfaction**: Satisfaction level with the workplace environment (1 to 4).
12. **Gender**: Gender of the employee (Male/Female).
13. **HourlyRate**: Hourly wage of the employee.
14. **JobInvolvement**: Level of involvement in their job (1 to 4).
15. **JobLevel**: Level of the job position (1 to 5).
16. **JobRole**: The employee's job role (e.g., Sales Executive, Research Scientist).
17. **JobSatisfaction**: Satisfaction with the job (1 to 4).
18. **MaritalStatus**: Marital status (e.g., Single, Married, Divorced).
19. **MonthlyIncome**: Monthly salary of the employee.
20. **MonthlyRate**: Monthly rate of pay.
21. **NumCompaniesWorked**: Number of companies the employee has worked at.
22. **Over18**: Whether the employee is over 18 (usually always "Yes").
23. **OverTime**: Whether the employee works overtime (Yes/No).
24. **PercentSalaryHike**: Percentage increase in salary.
25. **PerformanceRating**: Employee's performance rating (1 to 4).
26. **RelationshipSatisfaction**: Satisfaction with relationships at work (1 to 4).
27. **StandardHours**: Standard working hours (typically 40).
28. **StockOptionLevel**: Level of stock options granted (0 to 3).
29. **TotalWorkingYears**: Total years the employee has worked.
30. **TrainingTimesLastYear**: Number of training sessions attended last year.
31. **WorkLifeBalance**: Work-life balance rating (1 to 4).
32. **YearsAtCompany**: Total years the employee has been at the current company.
33. **YearsInCurrentRole**: Total years the employee has been in their current role.
34. **YearsSinceLastPromotion**: Years since the employee's last promotion.
35. **YearsWithCurrManager**: Total years the employee has worked with their current manager.
